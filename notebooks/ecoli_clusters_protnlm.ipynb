## Sequence data

### Reading data

Reading pre-processed Uniprot data for all organisms:

```python
from subpred.util import load_df

sequences = load_df("uniprot")
sequences
```

### Selecting organism

Filtering for *E. Coli* Strain K12: 

```python
sequences = sequences[sequences.organism_id == 83333]
sequences
```

### Filtering

Keeping proteins that have been manually reviewed, or have sequence evidence at protein level:

```python
sequences = sequences[(sequences.protein_existence == 1) | (sequences.reviewed)]
sequences
```

### Clustering:

The clustering at 70% sequence identity does not remove a lot of proteins for *E. coli*. This is different for other organisms. 

```python
from subpred.cdhit import cd_hit

clusters = cd_hit(sequences=sequences.sequence, identity_threshold=70)
sequences = sequences.loc[clusters]
sequences
```

## Annotation data

Next, we will load and filter the annotation datasets, such as Gene Ontology. This is because we need some of that data during feature generation (PSSMs were only calculated for transmembrane transporters).

**Load GO hierarchy:**

Get set of transmembrane transport GO terms. 

```python
from subpred.ontology import GeneOntology, GO_FILE

go_owl = GeneOntology(GO_FILE)
transmembrane_transport_go_terms = go_owl.get_descendants(go_owl.get_identifier("transmembrane transporter activity"))
```

**GO Uniprot annotations:**

Read pre-processed Uniprot GO annotations:

```python
go_annotations_uniprot = load_df("go")
go_annotations_uniprot
```

**Creating dataset of transporter GO term annotations**

GO annotations for the clustered sequence dataset from *E Coli.*

```python
# filter GO annotations for descendants of transmembrane transporter activity
go_annotations_uniprot = go_annotations_uniprot[go_annotations_uniprot.go_id.isin(transmembrane_transport_go_terms)].reset_index(drop=True)
go_annotations_uniprot
```

```python
# add ancestors of all remaining terms
go_annotations_uniprot = go_annotations_uniprot.assign(ancestors = go_annotations_uniprot.go_id.transform(lambda go_id: go_owl.get_ancestors(go_id)))
assert go_annotations_uniprot.apply(lambda row: row.go_id in row.ancestors, axis=1).all()
go_annotations_uniprot
```

```python
# add ancestor annotations
go_annotations_uniprot = go_annotations_uniprot.drop("go_id", axis=1).rename(columns={"ancestors": "go_id"}).explode("go_id")
go_annotations_uniprot
```

```python
# filter GO annotations and their ancestors for descendants of transmembrane transporter activity again
go_annotations_uniprot = go_annotations_uniprot[go_annotations_uniprot.go_id.isin(transmembrane_transport_go_terms)].reset_index(drop=True)
go_annotations_uniprot = go_annotations_uniprot.drop_duplicates().reset_index(drop=True)
go_annotations_uniprot
```

```python
# annotate with labels
go_annotations_uniprot = go_annotations_uniprot.assign(go_term = go_annotations_uniprot.go_id.transform(go_owl.get_label))
go_annotations_uniprot
```

## Feature generation

Here, we will use PSSM and protein embeddings features to encode the protein sequences into vectors of length *n*.

### ProtNLM

One of the most reliable features at the moment seems to be ProtNLM, which provides NLP embeddings for individual proteins. Swissprot has pre-calculated embeddings for all *E. Coli* proteins available for download. 

```python
import h5py
import pandas as pd
import numpy as np


def read_uniprot_embedding(h5_path: str):
    with h5py.File(h5_path, "r") as file:
        embeddings_dict = {
            sequence_id: np.array(embedding) for sequence_id, embedding in file.items()
        }
        return pd.DataFrame.from_dict(embeddings_dict, orient="index")


protnlm_embeddings_ecoli = read_uniprot_embedding(
    "/home/ad/protnlm_test/ecoli-per-protein.h5"
)
protnlm_embeddings_ecoli
```

**How many proteins in the dataset have embeddings available in the dataset?**

- More than 99% of proteins, that is good enough.

```python
# protnlm_embeddings_ecoli.loc[sequences.index]
ecoli_proteins = set(sequences.index.tolist())
available_embeddings = set(protnlm_embeddings_ecoli.index)

print(
    f"embeddings are available for {round(len(ecoli_proteins & available_embeddings) / len(ecoli_proteins), 3)}% of proteins in ecoli"
)
```

**Which proteins have no embeddings?**

The proteins with no embeddings available mostly seem to be related to the relaxasome, and bacterial conjugation/replication in general. There are no obvious transmembrane transporters among these excluded proteins:

```python
print("Embeddings not available for:")
sequences.loc[list(ecoli_proteins - available_embeddings)]
```

**Final feature dataset:**

```python
feature_protnlm = protnlm_embeddings_ecoli.loc[
    list(ecoli_proteins & available_embeddings)
]

feature_protnlm
```
