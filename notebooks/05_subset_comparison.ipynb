```python
df_uniprot_goa_all = (
    df_uniprot_goa[
        df_uniprot_goa.go_term_ancestor != "transmembrane transporter activity"
    ][["Uniprot", "go_term_ancestor"]]
    .drop_duplicates()
    .reset_index(drop=True)
)
df_uniprot_goa_all_proteins = set(df_uniprot_goa_all.Uniprot.unique())


def get_subset_stats(subset):
    df_uniprot_goa_subset = (
        df_uniprot_goa[df_uniprot_goa.go_id_ancestor.isin(subset)][
            ["Uniprot", "go_term_ancestor"]
        ]
        .drop_duplicates()
        .reset_index(drop=True)
    )
    display(df_uniprot_goa_subset.groupby("go_term_ancestor").agg(len))
    df_uniprot_goa_subset_proteins = set(df_uniprot_goa_subset.Uniprot.unique())
    print(
        len(df_uniprot_goa_all_proteins & df_uniprot_goa_subset_proteins),
        len(df_uniprot_goa_all_proteins),
    )
    print(
        len(df_uniprot_goa_all_proteins & df_uniprot_goa_subset_proteins)
        / len(df_uniprot_goa_all_proteins),
    )
```

```python
get_subset_stats(schaadt2010)
get_subset_stats(mishra2014)  # TODO use keywords
get_subset_stats(denger2022)
get_subset_stats(denger2024_high)
get_subset_stats(denger2024_low)
```
